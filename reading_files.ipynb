# Searching through a file

```python
# Searching through a file

file = open("spacex.txt")
for line in file:
    if line.startswith("Florida"):
        print(line)
```

# Searching through a file fixed

```python
file = open("spacex.txt")
for line in file:
    line = line.rstrip()
    if line.startswith("Florida"):
        print(line)
```

```python
# Skipping with continue
file = open("spacex.txt")
for line in file:
    line = line.rstrip()
    if not line.startswith("From:"):
        continue
    print(line)
```
