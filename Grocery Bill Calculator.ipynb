```python
# Grocery Bill Calculator
# A simple sequential program - no loops, no conditions

# Step 1: Define item prices
fruit_price = 350
bread_price = 7000
soup_price = 2000

# Step 2: Calculate subtotal
subtotal = fruit_price + bread_price + soup_price

# Step 3: Calculate tax (10%)
tax_rate = 0.10
tax = subtotal * tax_rate

# Step 4: Calculate total
total = subtotal + tax

# Step 5: Print the receipt
print("----- GROCERY RECEIPT -----")
print("Fruit:", fruit_price)
print("Bread:", bread_price)
print("Soup:", soup_price)
print("Subtotal:", subtotal)
print("Tax (10%):", tax)
print("Total:", total)
```
