```python
cart = ["120", "85", "abc", "200", "xx", "50", "9O"]

total = 0
bad_count = 0
good_count = 0

print("Let me check your cart items one by one...\n")

for item in cart:
    try:
        price = int(item)
        total = total + price
        good_count = good_count + 1
        print("Got it! Rs.", price, "added to your bill.")

    except:
        bad_count = bad_count + 1
        print("Hmm,", item, "does not look like a price. Skipping it!")

print("\nAlright, let me give you the final result!\n")
print("Valid items found   :", good_count)
print("Typos we skipped    :", bad_count)
print("Your total bill is  : Rs.", total)
print("\nThanks for shopping with us, have a great day!")
```
